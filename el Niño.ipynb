# Linear regression for el Niño dataset
<p>Gradient descent to find the line of the best fit that predicts relationship between air temperature and sea surface temperature during el Niño phenomena.</p>
<p>Some information can be found <a href='https://en.wikipedia.org/wiki/Gradient_descent'>here</a> and <a href='https://spin.atomicobject.com/2014/06/24/gradient-descent-linear-regression/'>here.</a> We will use [el Niño dataset](https://www.kaggle.com/uciml/el-nino-dataset/) from Kaggle.</p>

## Dependencies

```python
# Math library
import numpy as np
# Data processing library
import pandas as pd
# Plotting library
import matplotlib.pyplot as plt
%matplotlib inline
```

## Some useful variables

```python
# Number of gradient descent steps
num_iter = 1000
# Size of the training data and of the test data
batch_size = 1000
# Value that multiplies the gradient
learning_rate = 0.00001
```

## Reading and preparing the dataset

```python
#Reads the csv file wich contains the data
csv = pd.read_csv('elnino.csv')
csv
```

```python
#Gets two correlated columns: Air Temp and Sea Surface Temp
csv = csv.get([' Air Temp', ' Sea Surface Temp'])
csv
```

This dataset contains some rows with blank values (represented by .). We need to clean those rows.

```python
print('Length of dataset:', csv.shape[0])

#This function replaces '.' by np.nan to use in the function dropna and converts other values to float.
#It's necessary because '.' makes pandas read the values in the dataframe like a string.
def replace_by_nan(value):
    if value == '.':
        return np.nan
    else:
        return float(value)
    
# Using replace_by_nan() in Air Temp column.
csv[' Air Temp'] = csv[' Air Temp'].apply(replace_by_nan)
# Using replace_by_nan() in Sea Surface Temp column.
csv[' Sea Surface Temp'] = csv[' Sea Surface Temp'].apply(replace_by_nan)

#Cleaning the dataset
csv = csv.dropna(axis=0, how='any')

print('Length of cleaned dataset:', csv.shape[0])
```

Separated data for the training and the test.

```python
train_data = csv[:batch_size]
test_data = csv[(batch_size+1):(2*batch_size+1)]

print('Train data lenght:', train_data.shape[0])
print('Test data lenght:', test_data.shape[0])
```

## Plotting some points
<p>Plotting some for better visualization of the linearity correlation.</p>

```python
# Gets a sample of the entire dataset.
plot = csv.sample(100)
# x label
plt.xlabel('Air temperature')
# y label
plt.ylabel('Sea surface temperature')
# Plotting
plt.plot(plot[' Air Temp'], plot[' Sea Surface Temp'], 'ro')
plt.show()
```

## Gradient descent

<p>To measure the accuracy of our model we will use the <a href='https://en.wikipedia.org/wiki/Mean_squared_error'>mean squared error</a> formula.</p>
<img src='https://wikimedia.org/api/rest_v1/media/math/render/svg/67b9ac7353c6a2710e35180238efe54faf4d9c15'>

```python
#Point format is (x,y) or [x,y]
def meanSquaredError(m, b, points):
    total_error = 0
    for i in points.values:
        x = i[0]
        y = i[1]
        # error = (Y - (m*X + b))^2
        total_error += ((y - (m*x + b)) ** 2)
    return (total_error/float(len(points)))
```

<p>Gradient descent computes the partial derivative for m and b.</p>
<p><img src='https://spin.atomicobject.com/wp-content/uploads/linear_regression_gradient1.png'></p>

```python
def gradientDescent(m, b, learning_rate, points):
    m_grad = 0
    b_grad = 0
    n = float(len(points))
    
    #Computes the gradients
    for i in points.values:
        x = i[0]
        y = i[1]
        # Partial derivative for m
        m_grad += -(2/n) * x * (y - ((m*x) + b))
        # Partial derivative for b
        b_grad += -(2/n) * (y - (m*x) + b)
        
    # If we take too large of a step, we may step over the minimum.
    # However, if we take small steps, it will require many iterations to arrive at the minimum.
    # Learning rate prevents it.
    m -= (learning_rate * m_grad)
    b -= (learning_rate * b_grad)
    
    return m, b
```

```python
# Predicts y for a given x value
def pred(m, b, x):
    y = m*x + b
    return y

# Trains our model
def mathmagic(train_data, test_data,iterations, learning_rate):
    
    print('Starting', iterations, 'iterations.')
    
    # Initial m and b values
    m = 0
    b = 0
    
    # Training
    for i in range(iterations):
        
        # Error for some steps
        if(i % 100) == 0:
            error = meanSquaredError(m, b, test_data)
            print('Error after', i, 'iterations:', error)
            
        # Gradient descent
        m, b = gradientDescent(m, b, learning_rate, train_data)
        
    # Final error
    error = meanSquaredError(m, b, test_data)
    print('Error after', iterations, 'iterations:', error)
    
    return m,b
```

# Training step

```python
m,b = mathmagic(train_data, test_data, num_iter, learning_rate)
```

```python
sample = test_data.sample(10)
for i in sample.values:
    x = i[0]
    y = i[1]
    print('X value:', x)
    print('Y value:', y)
    print('Predicted Y value:', pred(m, b, x))
```

## Plotting!
